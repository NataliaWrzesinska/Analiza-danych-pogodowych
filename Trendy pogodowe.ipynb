### 4.3 Stworzenie skryptu do pozyskiwania danych z NOAA

Do pozyskiwania najświeższych danych pogodowych z serwisu NOAA (National Oceanic and Atmospheric Administration) wykorzystano specjalnie opracowany skrypt w języku Python. Skrypt ten automatycznie pobiera dane meteorologiczne dla stacji pogodowej w Polsce, umożliwiając dostęp do aktualnych informacji pogodowych.

```python
# Funkcja pobierająca dane ze strony NOAA
def pobierz_dane_noaa(id_stacji, rok_startowy):
    baza_url = "https://www.ncei.noaa.gov/access/services/data/v1"
    wszystkie_dane = []
    data_koncowa = datetime.now().strftime('%Y-%m-%d')  # Ustalanie daty końcowej na dzisiejszą datę

    # Pobieranie danych rok po roku
    for rok in range(rok_startowy, datetime.now().year + 1):
        parametry = {
            'dataset': 'daily-summaries',
            'stations': id_stacji,
            'startDate': f'{rok}-01-01',
            'endDate': data_koncowa if rok == datetime.now().year else f'{rok}-12-31',
            'format': 'csv',
            'units': 'metric'
        }
        odpowiedz = requests.get(baza_url, params=parametry)  # Wysłanie żądania do serwera
        if odpowiedz.status_code == 200:
            dane = pd.read_csv(StringIO(odpowiedz.text))  # Wczytanie danych do DataFrame
            wszystkie_dane.append(dane)  # Dodanie danych do listy
        else:
            print(f'Nie udało się pobrać danych dla roku {rok}')  # Informacja o niepowodzeniu

    polaczone_dane = pd.concat(wszystkie_dane, ignore_index=True)  # Połączenie wszystkich danych w jeden DataFrame
    return polaczone_dane

# Ustalenie parametrów
id_stacji = "PLM00012375"
rok_startowy = 2024

dane = pobierz_dane_noaa(id_stacji, rok_startowy)
dane.to_excel("dane_pogodowe_2024.xlsx", index=False)  # Zapisanie danych do pliku Excel
print("Dane zostały zapisane do pliku dane_pogodowe_2024.xlsx")  # Potwierdzenie zapisu danych
```

## 6. Ochrona danych

Celem ochrony danych jest zabezpieczenie ich przed nieautoryzowanym dostępem, utratą oraz uszkodzeniem. W kontekście analizy danych pogodowych, ochrona danych jest kluczowa dla zapewnienia ich integralności, poufności i dostępności.

### 6.1 Zabezpieczenie danych przed nieautoryzowanym dostępem

W dzisiejszych czasach ochrona danych przed nieautoryzowanym dostępem jest kluczowym elementem każdego projektu, zwłaszcza gdy dotyczy on wrażliwych informacji. W celu zabezpieczenia danych użyłam odpowiednich mechanizmów szyfrowania oraz kontroli dostępu. W tej części przedstawiam, jak zaszyfrować pliki z danymi oraz jak ustawić prawa dostępu do tych plików, aby zapewnić ich bezpieczeństwo.

```python
# Generowanie klucza szyfrowania
klucz = Fernet.generate_key()
cipher_suite = Fernet(klucz)

# Sciezki do plikow
sciezka_zaszyfrowane_dane = 'dane_pogodowe_zaszyfrowane.xlsx'
sciezka_klucz = 'klucz_szyfrowania.key'

# Funkcja do szyfrowania pliku
def zaszyfruj_plik(sciezka_wejsciowa, sciezka_wyjsciowa, cipher):
    with open(sciezka_wejsciowa, 'rb') as f:
        dane = f.read()
    zaszyfrowane_dane = cipher.encrypt(dane)
    with open(sciezka_wyjsciowa, 'wb') as f:
        f.write(zaszyfrowane_dane)

# Szyfrowanie pliku z danymi
zaszyfruj_plik(sciezka, sciezka_zaszyfrowane_dane, cipher_suite)

# Zapisanie klucza szyfrowania w bezpiecznym miejscu
with open(sciezka_klucz, 'wb') as f:
    f.write(klucz)

# Ustawienie praw dostepu do pliku (tylko wlasciciel moze czytac i zapisywac)
os.chmod(sciezka_zaszyfrowane_dane, stat.S_IRUSR | stat.S_IWUSR)
os.chmod(sciezka_klucz, stat.S_IRUSR | stat.S_IWUSR)

print("Plik został zaszyfrowany i zapisany jako 'dane_pogodowe_zaszyfrowane.xlsx'.")
print("Klucz szyfrowania został zapisany jako 'klucz_szyfrowania.key'.")
print("Ustawiono prawa dostepu do plikow.")
```

Zastosowanie szyfrowania danych oraz kontrola dostępu do plików pozwalają na skuteczne zabezpieczenie informacji przed nieautoryzowanym dostępem. Szyfrowanie za pomocą biblioteki cryptography zapewnia, że dane są chronione zarówno podczas przechowywania, jak i przesyłania. Ustawienie odpowiednich praw dostępu na poziomie systemu operacyjnego dodatkowo zwiększa bezpieczeństwo, ograniczając możliwość dostępu tylko do autoryzowanych użytkowników.

Podsumowując, zastosowanie tych metod w projekcie nie tylko chroni dane pogodowe, ale również spełnia współczesne standardy bezpieczeństwa danych, co jest niezbędne w każdej profesjonalnej analizie danych. Dalsze działania powinny obejmować regularny przegląd i aktualizację polityk bezpieczeństwa, aby uwzględniać nowe zagrożenia i technologie.

### 7.1 Dodanie kolumny z rokiem oraz kolumny z sezonem

Dodanie kolumny z rokiem oraz kolumny z sezonem umożliwia bardziej szczegółową analizę danych pogodowych, uwzględniając różnice sezonowe. Dzięki temu można łatwiej zidentyfikować sezonowe wzorce i trendy w danych.

```python
# Konwersja kolumny z datami na format daty
dane_pogodowe['data'] = pd.to_datetime(dane_pogodowe['data'])

# Dodanie kolumny z rokiem
dane_pogodowe['rok'] = dane_pogodowe['data'].dt.year

# Definicja funkcji do klasyfikacji sezonów
def get_season(month):
    if month in [12, 1, 2]:
        return 'zima'
    elif month in [3, 4, 5]:
        return 'wiosna'
    elif month in [6, 7, 8]:
        return 'lato'
    elif month in [9, 10, 11]:
        return 'jesień'

# Dodanie kolumny z sezonem
dane_pogodowe['sezon'] = dane_pogodowe['data'].dt.month.apply(get_season)

# Sprawdzenie pierwszych kilku wierszy danych
print(dane_pogodowe.head())
```

### 7.2 Analiza roczna i sezonowa

W celu zrozumienia długoterminowych zmian klimatycznych oraz identyfikacji wzorców pogodowych w Polsce, przeprowadzono szczegółową analizę roczną i sezonową danych pogodowych za okres 2014-2023. Analiza ta obejmuje zarówno średnie temperatury, jak i sumy opadów atmosferycznych oraz pokrywę śnieżną, co pozwala na wszechstronne zrozumienie dynamiki klimatycznej w różnych porach roku.

### 7.2.1 Średnie temperatury roczne

Celem analizy jest zidentyfikowanie trendów i zmienności temperatur, które mogą odzwierciedlać szersze zmiany klimatyczne.

```python
# Obliczenie średnich rocznych wartości temperatur
srednie_temperatury_roczne = dane_pogodowe.groupby('rok')['srednia_temp_c'].mean().reset_index(name='srednia_temp_roczna')
max_temperatury_roczne = dane_pogodowe.groupby('rok')['max_temp_c'].mean().reset_index(name='max_temp_roczna')
min_temperatury_roczne = dane_pogodowe.groupby('rok')['min_temp_c'].mean().reset_index(name='min_temp_roczna')

# Połączenie wyników w jedną tabelę
temperatury_roczne = srednie_temperatury_roczne.merge(max_temperatury_roczne, on='rok')
temperatury_roczne = temperatury_roczne.merge(min_temperatury_roczne, on='rok')

# Wyświetlenie wyników
print(temperatury_roczne)
```

```python
# Wizualizacja wyników
plt.figure(figsize=(10, 6))

# Wykres średnich temperatur rocznych
plt.plot(temperatury_roczne['rok'], temperatury_roczne['srednia_temp_roczna'], marker='o', linestyle='-', label='Średnia temperatura roczna')
# Wykres maksymalnych temperatur rocznych
plt.plot(temperatury_roczne['rok'], temperatury_roczne['max_temp_roczna'], marker='o', linestyle='-', label='Maksymalna temperatura roczna')
# Wykres minimalnych temperatur rocznych
plt.plot(temperatury_roczne['rok'], temperatury_roczne['min_temp_roczna'], marker='o', linestyle='-', label='Minimalna temperatura roczna')

# Dodanie tytułu i etykiet osi
plt.title('Średnie roczne wartości temperatur w latach 2014-2023')
plt.xlabel('Rok')
plt.ylabel('Temperatura (°C)')

# Dodanie legendy
plt.legend()

# Wyświetlenie wykresu
plt.show()
```

Zmienność temperatur:

- Z danych wynika, że zarówno średnia, maksymalna, jak i minimalna temperatura roczna wykazują pewną zmienność na przestrzeni lat.

Trendy temperatur:

- Maksymalne temperatury roczne generalnie wydają się wyższe w ostatnich latach z wyjątkiem spadków w niektórych latach. Średnie temperatury również wykazują tendencję wzrostową do 2023 roku, a minimalne temperatury mają tendencję do spadania w ostatnim czasie.

### 7.4 Analiza zmian klimatycznych

### 7.4.1 Analiza zmian klimatycznych metodą regresji liniowej

Metoda regresji liniowej jest techniką statystyczną służącą do modelowania zależności między zmienną zależną a jedną lub większą liczbą zmiennych niezależnych. W kontekście analizy klimatycznej, regresja liniowa pozwala na identyfikację trendów w danych pogodowych na przestrzeni lat. Współczynnik nachylenia (ang. slope) wskazuje na tempo zmiany zmiennej zależnej (np. temperatura) w odpowiedzi na zmiany zmiennej niezależnej (np. rok), natomiast punkt przecięcia (ang. intercept) to wartość zmiennej zależnej, gdy zmienna niezależna wynosi zero.

```python
# Filtracja danych, aby wykluczyć rok 2024
dane_pogodowe = dane_pogodowe[dane_pogodowe['rok'] < 2024]

# Obliczenie średnich rocznych wartości temperatur
srednie_temperatury_roczne = dane_pogodowe.groupby('rok')['srednia_temp_c'].mean().reset_index(name='srednia_temp_roczna')

# Przygotowanie danych do regresji
X = srednie_temperatury_roczne['rok'].values.reshape(-1, 1)
y = srednie_temperatury_roczne['srednia_temp_roczna'].values

# Tworzenie modelu regresji liniowej
model = LinearRegression()
model.fit(X, y)

# Przewidywanie wartości
y_pred = model.predict(X)

# Współczynnik nachylenia i punkt przecięcia
slope = model.coef_[0]
intercept = model.intercept_

# Rysowanie wykresu z regresją liniową
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='orange', label='Średnia temperatura roczna')
plt.plot(X, y_pred, color='green', linewidth=2, label='Regresja liniowa')
plt.xlabel('Rok')
plt.ylabel('Średnia temperatura (°C)')
plt.title('Trend średnich rocznych temperatur (2014-2023)')
plt.legend()
plt.grid(True)
plt.show()

# Wyświetlenie współczynnika nachylenia i punktu przecięcia
print(f'Współczynnik nachylenia: {slope}')
print(f'Punkt przecięcia: {intercept}')
```

```python
# Obliczenie rocznych sum opadów
suma_opadow_rocznie = dane_pogodowe.groupby('rok')['opady_mm'].sum().reset_index(name='suma_opadow_rocznie')

# Przygotowanie danych do regresji
X_opady = suma_opadow_rocznie['rok'].values.reshape(-1, 1)
y_opady = suma_opadow_rocznie['suma_opadow_rocznie'].values

# Tworzenie modelu regresji liniowej
model_opady = LinearRegression()
model_opady.fit(X_opady, y_opady)

# Przewidywanie wartości
y_opady_pred = model_opady.predict(X_opady)

# Współczynnik nachylenia i punkt przecięcia
slope_opady = model_opady.coef_[0]
intercept_opady = model_opady.intercept_

# Rysowanie wykresu z regresją liniową
plt.figure(figsize=(10, 6))
plt.scatter(X_opady, y_opady, color='blue', label='Suma opadów rocznie')
plt.plot(X_opady, y_opady_pred, color='green', linewidth=2, label='Regresja liniowa')
plt.xlabel('Rok')
plt.ylabel('Suma opadów (mm)')
plt.title('Trend rocznych sum opadów (2014-2023)')
plt.legend()
plt.grid(True)
plt.show()

# Wyświetlenie współczynnika nachylenia i punktu przecięcia
print(f'Współczynnik nachylenia: {slope_opady}')
print(f'Punkt przecięcia: {intercept_opady}')
```

Średnie roczne temperatury:

- Współczynnik nachylenia linii regresji wskazuje na wzrostowy trend w średnich rocznych temperaturach w latach 2014-2023. Jest to dowód na ocieplenie klimatu w Polsce.
- Chociaż wzrost temperatury jest widoczny, tempo wzrostu jest stosunkowo niewielkie, co może sugerować, że zmiany są stopniowe.
 
Roczne sumy opadów:

- Wyniki analizy regresji dla rocznych sum opadów wskazują na wzrostowy trend w ilości opadów w analizowanym okresie.
- Zwiększająca się suma opadów może wpływać na dostępność zasobów wodnych oraz warunki hydrologiczne w regionie, co ma istotne znaczenie dla zarządzania zasobami wodnymi i planowania działań adaptacyjnych.

Analiza czasowa przy użyciu regresji liniowej pozwala na identyfikację długoterminowych trendów w danych pogodowych. Wzrost średnich temperatur i rosnące sumy opadów mogą mieć istotne konsekwencje dla środowiska, gospodarki oraz codziennego życia. Regularne monitorowanie i analiza tych trendów są niezbędne dla adaptacji do zmieniających się warunków klimatycznych.

### 7.4.2 Analiza zmian klimatycznych metodą lasu losowego

Metoda lasu losowego (Random Forest) jest techniką uczenia maszynowego służącą do klasyfikacji i regresji. Polega na tworzeniu wielu drzew decyzyjnych podczas treningu i zwracaniu średniej przewidywań poszczególnych drzew w celu uzyskania dokładniejszych i bardziej stabilnych wyników. W tej analizie zastosowano las losowy do przewidywania średnich temperatur w Polsce na podstawie historycznych danych pogodowych.
